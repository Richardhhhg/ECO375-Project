```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("concatenated/df_all_features.csv")
```

```python
df.loc[df['Composite'].idxmax()]
```

```python
corr_matrix = df.drop(columns=['Country']).corr()
plt.figure(figsize=(20, 20))
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.show()
```

```python
df.drop(columns = ['Unnamed: 0', 'GDP_y'], inplace = True)
```

```python
df['log_gdp_capita'] = np.log(df['GDP_Per_Capita'])
```

```python
GDP_df = df[['log_gdp_capita', 'GDP_Per_Capita', 'GDP_Growth']]
```

```python
fig, ax = plt.subplots(1, 3, figsize = (24, 8))
ax = ax.flatten()

for i, column in enumerate(GDP_df.columns):
    sns.histplot(GDP_df[column], ax = ax[i])

plt.show()
```
